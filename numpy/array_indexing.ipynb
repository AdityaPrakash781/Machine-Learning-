# NUMPY INDEXING AND SLICING OF VECTORS <br>
Same as lists in python

```python
arr = np.arange(10,21)
print(arr)

#SLICING
arr_slice = arr[3::2]
print(arr_slice)
```

# INDEXING AND SLICING OF MATRIX

```python
matrix = np.arange(1,31).reshape(6,5)

print(matrix,"\n")

print(matrix[1],"\n") #Prints the row

print(matrix[2][3],"\n") #Prints the number at that index

print(matrix[1:3,2:4],"\n") #Slicing

print(matrix[:,3]) #Output is a vector
```

# BOOLEAN INDEXING

```python
arr =  np.arange(11,21)
print(arr)

bool_index = arr % 2 == 0 #Condition stating elements of the array are even
print(bool_index)

arr = arr[bool_index] #Index with True value are accepted
print(arr)
```
