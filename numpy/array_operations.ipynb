Arithmetic Operations

```python
import numpy as np
```

```python
arr1 =  np.array([0,1,2,3,4])
arr2 =  np.array([5,6,7,8,9])
```

```python
arr1+arr2
```

```python
arr1*arr2
```

```python
arr1**arr2
```

```python
arr1//arr2
```

```python
arr1/arr2
```

Broadcasting

```python
arr1 + 10
```

```python
arr3 = np.arange(1,26).reshape(5,5)

arr3 + 10
```

Deep and Shallow Copy

```python
slice = arr3[1]

slice = slice + 50
print(slice)

print(arr3) # NOTICE : The values in the arr3 from which slice was taken doesnt change , this is the concept of Shallow Copy . ONLY the sliced array is assigned a new location in the RAM , if the whole array is passed on , like this , arr4 = arr3 , then both are reffering to the same array.This is DEEP COPY.

#To pass on an array without referring to the same array , do this

arr4 = arr3.copy()
```
