```python
import numpy as np
```

Basic Array Creations

```python
arr = np.array([1,2,3,4])
print(arr)
```

```python
l = [1,2,3.5,'supp']
print(l)
larr = np.array(l)
print(larr)
```

Only same data type is stored in a numpy array ,
As u can see in the above example , An array with multiple data types was converted into a string.
An array containing integers and floating point numbers will be converted into an array of floating point numbers.

Below is an example of a 2D array , array with dimensions above 2 are tensors.

```python
normalList = [[1,2,3],[4,5,6],[7,8,9]]
print(normalList,"\n")
matrix = np.array(normalList)
print(matrix)
```

Random Generation Function


```python
arr = np.random.rand(10) #Between -1 and 1 , Normalization
print(arr)

arr1 = np.random.randn(10) #Between -3 and 3 , i.e. , Standardization
print(arr1)

arr2 = np.random.randint(10,30,5) #Hover over it to know more about it  
print(arr2)
```
